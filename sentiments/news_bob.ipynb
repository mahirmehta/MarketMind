```python
import pandas as pd

# Read the CSV file with the news data
news_file_path = 'sorted_grouped_data.csv'
df_news = pd.read_csv(news_file_path)

# Read the CSV file with the stock data
stock_file_path = 'BANKBARODA.NS.csv'
df_stock = pd.read_csv(stock_file_path)

# Convert the 'Date' columns to datetime type in both DataFrames
df_news['Date'] = pd.to_datetime(df_news['Date'])
df_stock['Date'] = pd.to_datetime(df_stock['Date'])

# Merge the two DataFrames based on the 'Date' column
merged_df = pd.merge(df_news, df_stock, on='Date', how='left')

# Display the merged DataFrame
print(merged_df)
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the VADER lexicon (do this once)
import nltk
nltk.download('vader_lexicon')
```

```python
# Create a SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Apply sentiment analysis to the 'Title' and 'Description' columns
merged_df['Title_Sentiment'] = merged_df['Title'].apply(lambda x: sia.polarity_scores(x)['compound'])
merged_df['Description_Sentiment'] = merged_df['Description'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Display the DataFrame with sentiment scores
print(merged_df[['Date', 'Title', 'Title_Sentiment', 'Description', 'Description_Sentiment']])
```
